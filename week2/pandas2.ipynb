```python
# Example Data for Diabetes

df = pd.DataFrame({"height": [146, 151, 167, 176, 190], "weight": [95, 55, 67, 92, 91], "diabetes": [True, False, False, False, True]})

df
```

---
<h4>Task 1</h4>
For every row, we want to create a new column called "obese" with these conditions:

1. if height < 180 and weight > 90
2. if weight > 90 and diabetes is true
---

```python
# Using what we know

is_obese = (df["height"] < 180) & (df["weight"] > 90)
is_obese2 = (df["weight"] > 90) & (df["diabetes"] == True)

df["obese"] = is_obese | is_obese2

df
```

```python
def is_obese(row):
  if row["height"] < 180 and row["weight"] > 90:
    return True
  elif row["weight"] > 90 and row["diabetes"] == "Yes":
    return True
  else:
    return False

df["obese"] = df.apply(is_obese, axis=1)
df
```

---
<h4>Task 2</h4>
For every row, we want to create a series called "means" that includs each column's means (if numeric) and mode (if binary)

---

```python
def mean_check(x):
    if x.dtype == "int64":
        return x.mean()
    if x.dtype == "bool":
        return x.mode()
means = df.apply(mean_check, axis=0)
means
```

<h4>Task 2</h4>
We want to create a dataframe with two rows: diabetes and non diabetes, each row must contain the mean in their respective group

---

```python
g = df.groupby("diabetes")

# creates a groupby object
```

```python
def means(x):
    return x.mean()

my_df = g.agg({"height": means, "weight": means, "obese": means})

my_df
```

```python
df2 = pd.DataFrame({"ID": [119, 121, 885, 635, 734, 85, 122, 152, 537], "country": ["US", "UK", "AUS", "NZ", "US", "US", "AUS", "NZ", "UK"], "score": [170, 110, 121, 155, 75, 66, 120, 117, 165], "bench": [120, 110, 135, 125, 121, 117, 122, 132, 136]})

df2
```

---
Task: Get a series that displays each country's mean bench and max score with respect to their group

---

```python
def means(x):
    return x.mean()

def maxs(x):
    return x.max()

g = df2.groupby("country")

result = g.agg({"score": maxs, "bench": means})

result
```
