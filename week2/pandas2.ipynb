```python
import pandas as pd
import numpy as np
```

```python
L1 = [2, 3, 1, 5] # ordinary python list
L2 = [6, 6, 2, 4]

NL1 = np.array(L1) # numpy array
NL2 = np.array(L2)

NL1 + NL1 ** 2 - (NL1 / NL2) # vectorised operations
```

```python
S1 = pd.Series([90, 60, 30, 40], index=["A", 3, "Z", 2])
S2 = pd.Series([70, 20, 30], index=["Z", 2, 9])

S1 + S2 # vectorised operations over series, resulting index automatically ordered alphanumerically
```

```python
S1["New"] = 10 # add new element
S1.drop("A", inplace=True) # remove element

S1
```

```python
print(S1.values) # access numpy array of values
print(S1.index) # access numpy array of indices (immutable)
```

```python
age = pd.Series([14, 16, 27, 33], index=[1, 2, 3, 4])
name = pd.Series(["Alice", "Bob", "Charlie", "David"], index=[1, 2, 3, 4])

df = pd.DataFrame({"Age": age, "Name": name}) # dataframe creation
df
```

```python
print(df.columns) # access list of columns 
print(df.index) # access list of indices
print(df.values) # prints matrix of values 
```

```python
df.loc[5] = [18, "Elsa"] # making a new row
df["country"] = ["Denmark", "USA", "Japan", "South Africa", "Argentina"] # making a new column
df
```

```python
df.drop(5, axis=0, inplace=True) # dropping a row
df.drop("country", axis=1, inplace=True) # dropping a column
df
```
